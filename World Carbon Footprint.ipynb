```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
```

```python
df=pd.DataFrame({
    'country':['China','United States','India','Russia','Japan',
               'Germany','Canada','South Korea',
               'Saudi Arabia','Brazil','South Africa','Australia',
              'United Kingdom','Nigeria','Indonesia','Iran',
              'Mexico','Turkey','Vietnam','Italy','Poland',
               'France','Thailand','Spain','Egypt','Argentina'],
    'iso_code':['CHN','USA','IND','RUS','JPN','DEU','CAN',
                'KOR','SAU','BRA','ZAF','AUS','GBR','NGA',
                'IDN','IRN','MEX','TUR','VNM','ITA','POL',
                'FRA','THA','ESP','EGY','ARG'],
    'region':['Asia','North America','Asia','Eurasia','Asia',
              'Europe','North America','Asia','Middle East',
              'South America','Africa','Oceania','Europe',
              'Africa','Asia','Middle East','North America',
              'Europe','Asia','Europe','Europe','Europe',
              'Asia','Europe','Africa','South America'],
    'total_co2':[13259,4682,2955,2069,944,582,575,573,
                 622,479,397,373,302,127,700,686,487,
                 481,327,322,321,315,282,254,265,184],
    'per_capita':[9.2,13.8,2.1,14.4,7.5,7.0,14.9,11.0,
                  17.1,2.2,6.5,14.2,4.2,0.6,2.5,7.7,
                  3.8,5.5,3.3,5.4,8.4,4.8,3.9,5.3,2.4,4.1]
    
})
df.head(10)
```

```python
fig=px.scatter(df,x='per_capita',y='total_co2',size='total_co2',
              color='region',
              hover_name='country',
               text='country',
               opacity=0.7,
              size_max=80,
              title="China's Carbon Dominance(2025)",
              labels={
                  'per_capita':"CO2 per Capita(Metric Tons)",
                  'total_co2':'Total CO2 Emission','country':'Country','region':'Region',
              })
fig.update_layout(template='plotly_dark',
                 title_font_size=20,
                  font_weight='bold',
                 margin=dict(l=50,r=50,t=80,b=50),
                 legend=dict(
                     title_text='Regions',
                     yanchor='top',
                     y=0.99,
                     xanchor='left',
                     x=1.02
                 ))
fig.update_yaxes(dtick=5,
                 showgrid=False,
                 visible=False
                )
fig.update_xaxes(showgrid=False)
fig.update_traces(textposition='top center',textfont=dict(size=10,color='white'),cliponaxis=False)
fig.show()
```
